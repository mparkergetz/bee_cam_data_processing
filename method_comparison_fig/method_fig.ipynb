```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('sample_method_counts.csv')

species_all = df['Species'].tolist()
counts_s1 = df['Camera'].tolist()
counts_s2 = df['BVT'].tolist()
counts_s3 = df['Net'].tolist()
```

```python
bar_width = 0.4  
species_all_reversed = list(reversed(species_all))
y_pos = range(len(species_all_reversed))  

fig, axs = plt.subplots(1, 3, figsize=(8.5, 3), sharey=True)

axs[0].barh(y_pos, list(reversed(counts_s1)), height=bar_width, color='m')
axs[0].set_yticks(y_pos)
axs[0].invert_yaxis()
axs[0].set_xlabel('Count')
axs[0].set_title('Camera Traps')

axs[1].barh(y_pos, list(reversed(counts_s2)), height=bar_width, color='c')
axs[1].set_yticks(y_pos)
axs[1].set_yticklabels([])  
axs[1].invert_yaxis()
axs[1].set_xlabel('Count')
axs[1].set_title('Blue Vane Traps')

axs[2].barh(y_pos, list(reversed(counts_s3)), height=bar_width, color='k')
axs[2].set_yticks(y_pos)
axs[2].set_yticklabels([]) 
axs[2].invert_yaxis()
axs[2].set_xlabel('Count')
axs[2].set_title('Hand Netting')

for ax in axs:
    ax.spines['left'].set_visible(False)
    ax.tick_params(axis='y', length=0)
    ax.set_xlim(0, 180)
    ax.set_xticks(range(0, 181, 30))

axs[0].set_yticklabels([f"{label}" for label in species_all_reversed], style='italic')
plt.tight_layout()
plt.savefig('method_comparison.svg', format='svg')
```
